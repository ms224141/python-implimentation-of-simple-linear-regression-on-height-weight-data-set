```python
iqr=2760
q3=6360
q1=3600
lowerbound=q3-1.5*iqr
upperbound=q1+1.5*iqr
```

```python
outlier_indices = df.index[df["area"].apply(lambda value: value < lowerbound or value > upperbound)].tolist()
```

```python
outlier_indices
```

```python
df=df.drop(outlier_indices)
```

```python
x = df[["bedrooms"]]  # Use double brackets to select the column as a DataFrame
y = df["price"]

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)

from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)  # This should now work without errors
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```
