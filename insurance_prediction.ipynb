```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/insurance.csv")
df
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
#  now we will check for outliers in it
df.describe()
```

```python
#  now check if there are any outlier in the age column or not
q1=27
q3=51
iqr=q3-q1
lb=q1-(1.5*iqr)
ub=q3+(1.5*iqr)
print(lb,ub)

# Use a single boolean mask to filter rows based on age
outliers = df[(df["age"] < lb) | (df["age"] > ub)]
outliers
```

```python
df.drop(outliers.index).reset_index(drop=True)
```

```python
x=df.drop(columns=["charges"])
y=df["charges"]
```

```python
from sklearn.feature_selection import RFE
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
x = pd.get_dummies(x, columns=['sex', 'smoker', 'region'], drop_first=True)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=15)
```

```python
#  first we have to convert sex into numaric
```

```python
recursion=RFE(estimator=LinearRegression(),n_features_to_select=3)
recursion.fit(x_train,y_train)
```

```python
y_pred=recursion.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```
